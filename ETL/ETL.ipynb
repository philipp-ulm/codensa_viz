# Pre-procesamiento de datos

```python
import pandas as pd
import numpy as np
```

```python
df_consumo = pd.read_csv("db_Uniandes.csv",sep = '|', header = 0, encoding = 'utf8')
clientes = pd.read_csv("Clientes.csv",sep = ';', header = 0, encoding = 'utf8')
tarifa = pd.read_csv("Tarifa.csv",sep = ';', header = 0, encoding = 'utf8')
```

```python
df_consumo[['CLIENTE','Fecha']] = df_consumo[['CLIENTE','Fecha']].astype(str)
```

```python
df2 = pd.merge(clientes, tarifa, how='left', left_on=['Tarifa','Estrato'], right_on=['Tarifa','Estrato'])
df2[['CLIENTE','Vigencia','Periodo']] = df2[['CLIENTE','Vigencia','Periodo']].astype(str)
```

```python
df = pd.DataFrame(df_consumo.groupby(['CLIENTE','Fecha'])['value'].sum().reset_index(name='Consumo'))
df['Consumo'] = df['Consumo']/1000 
df.head()
```

```python
df['Periodo'] = df['Fecha'].str.slice(0, 4) +  df['Fecha'].str.slice(4, 6)
df.head()
```

```python
df3 = pd.merge(df, df2, how='left', left_on=['CLIENTE','Periodo'], right_on=['CLIENTE','Periodo'])
df3 = df3.sort_values(['CLIENTE', 'Fecha'], ascending=[1, 1])
df3.shape
```

```python
df3 = df3[~df3['Vigencia'].isna()]
df3.shape
```

```python
df3.to_csv('Base_final_clientes.csv',index=False)
```
